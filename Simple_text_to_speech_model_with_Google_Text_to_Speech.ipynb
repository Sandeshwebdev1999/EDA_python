```python
from gtts import gTTS

text_to_say = "Hi, Buddy what is your name"

language = "en"

gtts_object = gTTS(text = text_to_say,
                  lang = language,
                  slow = True)

gtts_object.save("/content/gtts.wav")
```

 **`from gtts import gTTS`**
   ➤ Imports the **Google Text-to-Speech** module.

 **`text_to_say = "Hi, Buddy what is your name"`**
   ➤ The **text you want to convert to speech**.

 **`language = "en"`**
   ➤ Sets the **language** for speech (here, `"en"` for English).

 **`gtts_object = gTTS(text=text_to_say, lang=language, slow=True)`**
   ➤ Converts the text into a **voice object**.
   ➤ `slow=True` means the voice will be spoken slowly.

 **`gtts_object.save("/content/gtts.wav")`**
   ➤ Saves the generated voice as a **`.wav` audio file**.
