```python
# Step 1: Imports
!pip install ultralytics
from IPython.display import display, Javascript, clear_output
from google.colab.output import eval_js
from base64 import b64decode
from PIL import Image
import numpy as np
import cv2
import time
from ultralytics import YOLO
```

_____

```python
# Step 2: Load YOLOv8 model
model = YOLO("yolov8n.pt")
```

#### ‚úÖ What This Step Does:

- **Loads the Pre-trained YOLOv8 Model:**  
  We use the `YOLO` class from the Ultralytics library to load the **YOLOv8n (nano)** model, which is a lightweight and fast version of YOLOv8.

- **Model File ‚Äì `yolov8n.pt`:**  
  This is a pre-trained PyTorch model checkpoint provided by Ultralytics. It is trained on the **COCO dataset**, which can detect 80 common object classes like person, car, dog, etc.

#### üí° Why YOLOv8n?

- ‚ö° **Fast and Efficient**: Ideal for real-time object detection in resource-constrained environments like Google Colab.
- üß† **Pre-trained**: No additional training required for basic object detection.
- üß∞ **Plug-and-play**: Easily integrates with the detection pipeline that uses webcam input.


_____

```python
# Step 3: JavaScript to access webcam & capture frame
def capture_frame():
  display(Javascript('''
    async function captureImage() {
      const video = document.createElement('video');
      video.setAttribute("playsinline", "true");  // required to tell iOS safari we don't want fullscreen
      const stream = await navigator.mediaDevices.getUserMedia({ video: true });
      document.body.appendChild(video);
      video.srcObject = stream;
      await video.play();

      await new Promise(resolve => setTimeout(resolve, 500)); // wait for camera

      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      const context = canvas.getContext('2d');
      context.drawImage(video, 0, 0, canvas.width, canvas.height);
      stream.getTracks().forEach(track => track.stop());
      video.remove();
      return canvas.toDataURL('image/jpeg');
    }
    captureImage();
  '''))
  return eval_js("captureImage()")
```

#### ‚úÖ What This Step Does:

- **Integrates JavaScript into Colab:**  
  Uses JavaScript inside Google Colab to **access the user's webcam** and **capture a single frame**.

- **Steps in the JavaScript Logic:**
  - Creates a video element and requests webcam access via `navigator.mediaDevices.getUserMedia()`.
  - Displays the live video stream and waits briefly for the camera to initialize.
  - Captures a frame from the video and draws it on a canvas.
  - Converts the canvas image to a **Base64-encoded JPEG** string.
  - Stops the webcam stream and removes the video element from the DOM.

- **Returns Captured Frame:**  
  The function returns the captured image in Base64 format, which will be processed in the next step.

#### üí° Why Use JavaScript in Colab?

- Google Colab doesn‚Äôt natively support webcam access using Python alone.
- JavaScript provides browser-based media device access, making it possible to capture real-time images for processing.


______

```python
# Step 4: Convert JS image to OpenCV format
def js_to_image(js_data):
  image_bytes = b64decode(js_data.split(',')[1])
  np_array = np.frombuffer(image_bytes, dtype=np.uint8)
  return cv2.imdecode(np_array, cv2.IMREAD_COLOR)
```

#### ‚úÖ What This Step Does:

- **Purpose:**  
  Converts the Base64-encoded image (captured using JavaScript from the webcam) into an image format compatible with **OpenCV** for further processing and object detection.

#### üß† Step-by-Step Breakdown:

1. **`b64decode(js_data.split(',')[1])`**  
   - Extracts and decodes the image portion from the Base64 string received from the JavaScript webcam capture.
   - Base64 string format: `"data:image/jpeg;base64,<actual-encoded-data>"`

2. **`np.frombuffer(image_bytes, dtype=np.uint8)`**  
   - Converts the decoded bytes into a NumPy array of type `uint8`, which is a suitable format for image data.

3. **`cv2.imdecode(np_array, cv2.IMREAD_COLOR)`**  
   - Decodes the NumPy byte array into a BGR image using OpenCV, ready for processing or visualization.

#### üí° Why This Step Is Important:

- Ensures smooth transition from **webcam-captured JavaScript image** to **Python-processed OpenCV format**.
- Converts the raw image into a format that can be fed directly into YOLOv8 for object detection.

-----

```python
# Step 5: YOLOv8 processing function
def process_with_yolo(frame):
  results = model(frame, verbose=False)[0]
  return results.plot()
```

#### ‚úÖ What This Step Does:

- **Feeds the Captured Frame into YOLOv8:**  
  The function takes a single image frame (in OpenCV format) and runs it through the loaded YOLOv8 model for object detection.

#### üß† Step-by-Step Breakdown:

1. **`model(frame, verbose=False)[0]`**  
   - Passes the input frame to the YOLOv8 model for inference.
   - `verbose=False` disables logging output for a cleaner run.
   - The `[0]` accesses the first result from the model‚Äôs output (as YOLO returns a list of results).

2. **`results.plot()`**  
   - Plots the detection results on the original image.
   - The output image includes **bounding boxes, class labels**, and **confidence scores** for each detected object.

#### üéØ Output:

- Returns an annotated image (NumPy array) with all detected objects drawn using YOLOv8‚Äôs built-in visualization.

#### üí° Why This Step Is Crucial:

- This is the **core detection logic** of the project.
- It leverages YOLOv8's power to detect objects in real-time from webcam input and prepares the output for final display.


------

```python
# Step 6: Live loop for simulated video
def start_live_detection(frame_count=15, delay_sec=1):
  for i in range(frame_count):
    print(f"üîÅ Processing Frame {i+1}...")
    js_data = capture_frame()
    frame = js_to_image(js_data)
    processed = process_with_yolo(frame)
    display(Image.fromarray(processed))
    clear_output(wait=True)
    time.sleep(delay_sec)  # Optional: Adjust for smoother flow

# Step 7: RUN IT
start_live_detection(frame_count=15, delay_sec=1)
```

#### ‚úÖ What These Steps Do:

- **Simulates Real-Time Object Detection in Google Colab:**
  Runs a live loop to capture webcam frames, process them using YOLOv8, and display results frame by frame‚Äîmimicking a live video feed.

#### üß† Step-by-Step Breakdown:

1. **Loop over a number of frames (`frame_count`)**
   - Default is `15` frames with a `1-second` delay between each.
   - This simulates a live video experience in a notebook environment.

2. **`capture_frame()`**
   - Captures a fresh webcam frame using JavaScript.

3. **`js_to_image(js_data)`**
   - Converts the Base64 image into OpenCV format.

4. **`process_with_yolo(frame)`**
   - Runs object detection on the frame and adds bounding boxes and labels.

5. **`display(Image.fromarray(processed))`**
   - Displays the processed image with detected objects.

6. **`clear_output(wait=True)`**
   - Clears the output cell before displaying the next frame to create a fluid, live effect.

7. **`time.sleep(delay_sec)`**
   - Adds a small delay to control the pace of frame updates.
